```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Load all data


X_left_pd = np.load("PD_X_left.npy", allow_pickle=True)
X_right_pd = np.load("PD_X_right.npy", allow_pickle=True)
y = np.load("PD_y.npy", allow_pickle=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def visualize_patient_comparison(X_left_pd, X_right_pd, healthy_idx=0, pd_idx=-5):
    """
    Compare one 'Healthy' patient vs one 'PD' patient
    using both left and right wrist signals.
    """

    # Convert lists to arrays if needed
    X_left_pd = np.array(X_left_pd, dtype=object)
    X_right_pd = np.array(X_right_pd, dtype=object)

    # Extract patients
    healthy_left = X_left_pd[healthy_idx]
    healthy_right = X_right_pd[healthy_idx]
    pd_left = X_left_pd[pd_idx]
    pd_right = X_right_pd[pd_idx]

    # Sensor names (adjust if needed)
    sensor_names = ['v1', 'Accel_X', 'Accel_Y', 'Accel_Z', 'Gyro_X', 'Gyro_Y', 'Gyro_Z']

    fig = plt.figure(figsize=(20, 12))

    # --- LEFT WRIST COMPARISON ---
    for i in range(1, 7):  # Skip v1 (index 0)
        ax = plt.subplot(2, 6, i)
        ax.plot(healthy_left[:, i], label='Healthy', color='blue', linewidth=2)
        ax.plot(pd_left[:, i], label='PD', color='red', linewidth=2, alpha=0.7)
        ax.set_title(f'Left Wrist - {sensor_names[i]}', fontweight='bold', fontsize=10)
        ax.set_xlabel('Timestep')
        ax.set_ylabel('Value')
        ax.grid(True, alpha=0.3)
        if i == 1:
            ax.legend()

    # --- RIGHT WRIST COMPARISON ---
    for i in range(1, 7):
        ax = plt.subplot(2, 6, 6 + i)
        ax.plot(healthy_right[:, i], label='Healthy', color='blue', linewidth=2)
        ax.plot(pd_right[:, i], label='PD', color='red', linewidth=2, alpha=0.7)
        ax.set_title(f'Right Wrist - {sensor_names[i]}', fontweight='bold', fontsize=10)
        ax.set_xlabel('Timestep')
        ax.set_ylabel('Value')
        ax.grid(True, alpha=0.3)
        if i == 1:
            ax.legend()

    plt.suptitle("Healthy vs PD (Raw Signal Comparison) for movement: Relaxed", fontsize=18, fontweight='bold', y=1.02)
    plt.tight_layout()
    plt.subplots_adjust(top=0.9)
    plt.show()

# Run the comparison
visualize_patient_comparison(X_left_pd, X_right_pd)
```
